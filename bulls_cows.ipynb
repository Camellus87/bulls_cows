```python
import random

# def greet_user():
    # Vypíše uvítací zprávu a pravidla hry

def generate_secret_number():
    # Vytvoří náhodné 4ciferné číslo s unikátními číslicemi, nezačíná nulou
    digits = list("123456789")
    secret_number = random.choice(digits)  # Vybere první číslici, která nesmí být 0
    digits.remove(secret_number)  # Odstraní vybranou číslici
    digits.append("0")  # Přidá 0 zpět pro další volby
    secret_number += ''.join(random.sample(digits, 3))  # Vybere zbývající 3 číslice
    return secret_number

def validate_guess(guess):
    # Ověří, zda je uživatelský vstup validní
    if len(guess) != 4:
        return False, "The number must be exactly 4 digits long."
    if not guess.isdigit():
        return False, "The number must contain only digits."
    if guess[0] == '0':
        return False, "The number must not start with 0."
    if len(set(guess)) != 4:  # Kontroluje, zda jsou všechny číslice unikátní
        return False, "The number must not contain duplicate digits."
    return True, ""

def evaluate_guess(secret, guess):
    # Vyhodnotí počet bulls a cows
    bulls = sum(s == g for s, g in zip(secret, guess))
    cows = sum(s in guess and s != g for s, g in zip(secret, guess))
    return bulls, cows

def format_result(bulls, cows):
    # Vrátí formátovaný textový výstup
    bull_text = "bull" if bulls == 1 else "bulls"
    cow_text = "cow" if cows == 1 else "cows"
    return f"{bulls} {bull_text}, {cows} {cow_text}"

def get_performance_message(attempts):
    # Určí úroveň výkonu hráče
    if attempts <= 5:
        return "amazing"
    elif attempts <= 10:
        return "average"
    else:
        return "not so good"

#def main():
    # Hlavní logika hry

print("Hi there!")
print("-----------------------------------------------")
print("I've generated a random 4 digit number for you.")
print("Let's play a bulls and cows game.")
print("-----------------------------------------------")

secret_number = generate_secret_number()
attempts = 0
    
while True:
    guess = input("Enter a  4 digit number (or type exit to end): ")
    if guess.lower() == "exit":
            print("Hra ukončena. Vzdal jsi to.")
            break
    print("-----------------------------------------------")        
    is_valid, error_message = validate_guess(guess)
    if not is_valid:
        print(error_message)
        print("-----------------------------------------------")
        continue        
    attempts += 1
    bulls, cows = evaluate_guess(secret_number, guess)
    result = format_result(bulls, cows)
    print(result)
    print("-----------------------------------------------")        
    if bulls == 4:
        performance_message = get_performance_message(attempts)
        print(f"Correct, you've guessed the right number in {attempts} guesses!")
        print(f"That's {performance_message}!")
        print("-----------------------------------------------")
        break

#if __name__ == "__main__":
    # Spustí hlavní funkci při přímém spuštění skriptu
#    main()
```
