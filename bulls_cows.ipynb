```python
## projekt z engeta bulls and cows, finále v1.0
'''
bulls_cows: druhy projekt pro Engeto Python; v1.0

author: Kamil Hnátek
email: kamil.hnatek@gmail.com
discord: irenicus87 
'''
import random

def generovani_cisla():
    # Vytvoří náhodné 4ciferné číslo s unikátními číslicemi, nezačíná nulou
    digits = list("123456789")
    tajne_cislo = random.choice(digits)  # Vybere první číslici, která nesmí být 0
    digits.remove(tajne_cislo)  # Odstraní vybranou číslici
    digits.append("0")  # Přidá 0 zpět pro další volby
    tajne_cislo  += ''.join(random.sample(digits, 3))  # Vybere zbývající 3 číslice
    return tajne_cislo

def overeni_vstupu(hadej):
    # Ověří, zda je uživatelský vstup validní
    if len(hadej) != 4:
        return False, "The number must be exactly 4 digit long."
    if not hadej.isdigit():
        return False, "Use only digits!"
    if hadej[0] == '0':
        return False, "The number must not start with 0."
    if len(set(hadej)) != 4:  # Kontroluje, zda jsou všechny číslice unikátní
        return False, "Do not use duplicate digits, one number = 1x."
    return True, ""

def pocet_bulls_cows(cislo, hadej):
    # Vyhodnotí počet bulls a cows
    bulls = sum(s == g for s, g in zip(cislo, hadej))
    cows = sum(s in hadej and s != g for s, g in zip(cislo, hadej))
    return bulls, cows

def format_vystupu(bulls, cows):
    # Vrátí formátovaný textový výstup
    bull_text = "bull" if bulls == 1 else "bulls"
    cow_text = "cow" if cows == 1 else "cows"
    return f"{bulls} {bull_text}, {cows} {cow_text}"

def statistika(attempts):
    # Určí úroveň výkonu hráče
    if attempts <= 6:
        return "amazing"
    elif attempts <= 12:
        return "average"
    else:
        return "not so good"

# Hlavní logika hry
# Vypíše uvítací zprávu a pravidla hry
print("Hi there!")
print("-----------------------------------------------")
print("I've generated a random 4 digit number for you.")
print("Let's play a bulls and cows game.")
print("-----------------------------------------------")
cislo = generovani_cisla()
print(cislo)
attempts = 0

while True:
    hadej = input("Enter a  4 digit number (or type 'exit' to end): ")
    if hadej.lower() == "exit":
            print("Game ended by you. So sad :(")
            break
    print("-----------------------------------------------")        
    is_valid, error_message = overeni_vstupu(hadej)
    if not is_valid:
        print(error_message)
        print("-----------------------------------------------")
        continue        
    attempts += 1
    bulls, cows = pocet_bulls_cows(cislo, hadej)
    result = format_vystupu(bulls, cows)
    print(result)
    print("-----------------------------------------------")        
    if bulls == 4:
        performance_message = statistika(attempts)
        print(f"Correct, you've guessed the right number {hadej} in {attempts} attempts!")
        print(f"That is {performance_message}!")
        print("-----------------------------------------------")
        break
#konec
```
